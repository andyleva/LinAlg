2. –ù–∞–π—Ç–∏ –Ω–æ—Ä–º—ã –≤–µ–∫—Ç–æ—Ä–æ–≤  (4,2,4)  –∏  (12,3,4)  –∏ —É–≥–æ–ª –º–µ–∂–¥—É –Ω–∏–º–∏.

- $p = 1$ –ø–æ–ª—É—á–∞–µ–º $\ell_{1}$ –Ω–æ—Ä–º—É (_–º–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫–∞—è –Ω–æ—Ä–º–∞_, –∏–ª–∏ _–º–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ_)

$$\left\|x\right\|_{1} = \sum_{i}|x_{i}|;$$

- $p = 2$ –ø–æ–ª—É—á–∞–µ–º $\ell_{2}$ –Ω–æ—Ä–º—É (_–µ–≤–∫–ª–∏–¥–æ–≤–∞ –Ω–æ—Ä–º–∞_, –Ω–∞–∏–±–æ–ª–µ–µ —á–∞—Å—Ç–æ –∏—Å–ø–æ–ª—å–∑—É–µ–º–∞—è)

$$\left\|x\right\|_{2} = \sqrt{\sum_{i}|x_{i}|^{2}}.$$

—Ä–∞—Å—á–µ—Ç —É–≥–ª–∞ –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏
$$cos\varphi = \frac{(x,y)}{\left \|x\right \|\cdot\left \|y\right \|}.$$

```python
from numpy.linalg import norm
import math
a = np.array([4,2,4])
b = np.array([12,3,4])
print(f'–º–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫–∞—è –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ a: {norm(a, ord = 1)}')
print(f'–º–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫–∞—è –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ b: {norm(b, ord = 1)}')
print(f'–µ–≤–∫–ª–∏–¥–æ–≤–∞ –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ a: {norm(a, ord = 2)}')
print(f'–µ–≤–∫–ª–∏–¥–æ–≤–∞ –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ b: {norm(b, ord = 2)}')
cos_phi = np.dot(a,b) / norm(a) / norm(b)
print(f'—É–≥–æ–ª –º–µ–∂–¥—É a –∏ b: {math.degrees(cos_phi)} –≥—Ä–∞–¥—É—Å–æ–≤')
```

3. –ë—É–¥–µ—Ç –ª–∏ –ª–∏–Ω–µ–π–Ω–æ–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ –µ–≤–∫–ª–∏–¥–æ–≤—ã–º, –µ—Å–ª–∏ –∑–∞ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –ø—Ä–∏–Ω—è—Ç—å:

–±) —É—Ç—Ä–æ–µ–Ω–Ω–æ–µ –æ–±—ã—á–Ω–æ–µ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤?

__4.__ –ö–∞–∫–∏–µ –∏–∑ –Ω–∏–∂–µ–ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ –æ–±—Ä–∞–∑—É—é—Ç –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –±–∞–∑–∏—Å –≤ –ª–∏–Ω–µ–π–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ $\mathbb{R}^{3}$:<br>
–∞) $(1,0,0),(0,0,1);$ <br>
–±) $(1/\sqrt{2},-1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0), (0,0,1);$<br>
–≤) $(1/2, -1/2, 0), (0, 1/2, 1/2), (0,0,1);$<br>
–≥) $(1,0,0),(0,1,0),(0,0,1)?$ 

–ë–∞–∑–∏—Å e1,e2,‚Ä¶,en –µ–≤–∫–ª–∏–¥–æ–≤–∞ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–º, –µ—Å–ª–∏ –µ–≥–æ –≤–µ–∫—Ç–æ—Ä—ã –ø–æ–ø–∞—Ä–Ω–æ –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã –∏ –¥–ª–∏–Ω–∞ –∫–∞–∂–¥–æ–≥–æ –∏–∑ –Ω–∏—Ö —Ä–∞–≤–Ω–∞ –µ–¥–∏–Ω–∏—Ü–µ:
$$(ei,ej)= \begin{cases}
    1,i=j,\\
    0,i‚â†j\\
     \end{cases}$$
 $$, –≥–¥–µ  i = 1,2,‚Ä¶,n,  j = 1,2,‚Ä¶,n$$

–í –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –±–∞–∑–∏—Å–µ e1,e2,‚Ä¶,en –¥–ª–∏–Ω–∞ –≤–µ–∫—Ç–æ—Ä–∞ x –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ 
$$|x|=\sqrt{x1^2+x2^2+‚Ä¶+xn^2},$$
–≥–¥–µ x1,‚Ä¶,xn ‚Äî –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –≤–µ–∫—Ç–æ—Ä–∞ x

```python
def len_vector(a,b,c):
    a1 = a**2
    len_a = math.sqrt(a1.sum())
    b1 = b**2
    len_b = math.sqrt(b1.sum())
    c1 = c**2
    len_c = math.sqrt(c1.sum())
    return len_a,len_b,len_c
```

```python
print(f'–¥–ª–∏–Ω—ã –≤–µ–∫—Ç–æ—Ä–æ–≤ –±): {len_vector(np.array([math.sqrt(1/2),-math.sqrt(1/2),0]), np.array([math.sqrt(1/2),math.sqrt(1/2),0]), np.array([0,0,1]))}')
print(f'–¥–ª–∏–Ω—ã –≤–µ–∫—Ç–æ—Ä–æ–≤ –≤): {len_vector(np.array([1/2,-1/2,0]),np.array([0,1/2,1/2]),np.array([0,0,1]))}')
print(f'–¥–ª–∏–Ω—ã –≤–µ–∫—Ç–æ—Ä–æ–≤ –≥): {len_vector(np.array([1,0,0]),np.array([0,1,0]),np.array([0,0,1]))}')
```
